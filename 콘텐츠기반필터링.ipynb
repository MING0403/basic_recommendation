## 2. 데이터 전처리

## 3. 장르 콘텐츠 기반 유사도
- [a, b, c]로 구성되어 있는 장르별 유사도는 장르를 문자열로 변경한 후 **CountVecotrizer로 피처를 벡터화**해서 비교

```python
# !pip install --upgrade pip
# !pip install scikit-learn
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
movies_df['genres_literal'] = movies_df['genres'].apply(lambda x: (' ').join(x))
count_vect = CountVectorizer(min_df=0.0, ngram_range=(1,2))
genre_mat = count_vect.fit_transform(movies_df['genres_literal'])
```

```python
# movies_df.head()
```

```python
genre_mat.shape
# 4803개의 레코드와 276개의 개별 단어로 구성
```

- Cosine 유사도를 통해 장르 유사도 값 측정

```python
from sklearn.metrics.pairwise import cosine_similarity

genre_sim = cosine_similarity(genre_mat, genre_mat)
print(genre_sim.shape)
print(genre_sim[:5])
```

```python
# 유사도 높은 순으로 출력
genre_sim_sorted_ind = genre_sim.argsort()[:, ::-1]
genre_sim_sorted_ind[:1]
```

## 4. 유사도 기반 추출

```python
def sim_movie(df, sorted_ind, title_name, top_n=10):
    title_movie = df[df['title'] == title_name]
    
    # title명을 가진 영화의 인덱스 추출
    title_index = title_movie.index.values
    # genre_sim_sorted_ind에서 유사도 순으로 top n개 정렬
    sim_index = sorted_ind[title_index, :(top_n)]
    
    print(sim_index)
    sim_index = sim_index.reshape(-1)
    
    return df.iloc[sim_index]
```

```python
similar_movies = sim_movie(movies_df, genre_sim_sorted_ind, 'The Godfather', 10)
similar_movies[['title', 'vote_average']]
```

## 5. 가중평점
- IMDB의 가중평점의 공식 참고

```python
c = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(0.6) # 상위 60%에 해당하는 횟수
print('전체 영화평점:', round(c,3))
print('전체 투표횟수의 상위 60%:', round(m,3))
```

```python
percentile=0.6
c = movies_df['vote_average'].mean()
m = movies_df['vote_count'].quantile(percentile) # 상위 60%에 해당하는 횟수

def weighted_vote_avg(record):
    v = record['vote_count']
    R = record['vote_average']
    
    return ((v/(v+m))*R) + (((m/m+v))*c)
```

```python
movies_df['weighted_vote'] = movies_df.apply(weighted_vote_avg, axis=1)
```

```python
movies_df[['title', 'vote_average','weighted_vote','vote_count']].sort_values('weighted_vote', ascending=False)[:10]
```

## 6. 대부와 유사한 영화를 콘텐츠 기반 필터링 추천

```python
def find_sim_movies(df, sorted_ind, title_name, top_n=10):
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values
    
    # top n개의 장르 유사성이 높은 인덱스 추출
    sim_index = sorted_ind[title_index, :(top_n*2)]
    sim_index = sim_index.reshape(-1)
    
    # 기준 영화 인덱스는 제외
    sim_index = sim_index[sim_index != title_index]
    
    return df.iloc[sim_index].sort_values('weighted_vote', ascending=False)[:top_n]
```

```python
result = find_sim_movies(movies_df, genre_sim_sorted_ind, 'The Godfather', 10)
result[['title', 'vote_average', 'weighted_vote']]
```
